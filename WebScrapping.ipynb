```python

from bs4 import BeautifulSoup
import requests
import smtplib
import time
import datetime
from datetime import datetime, timedelta
```

```python

def get_current_date_time():
    now = datetime.now()
    
    # Subtrai 3 horas da hora atual
    adjusted_time = now - timedelta(hours=3)
    
    current_date = adjusted_time.strftime("%d/%m/%Y")
    current_time = adjusted_time.strftime("%H:%M:%S")
    
    return current_date, current_time

def get_usd_value():
    url = "https://economia.uol.com.br/cotacoes/cambio/"
    response = requests.get(url)
    response.raise_for_status()
    soup = BeautifulSoup(response.content, 'html.parser')
    usd_value_element = soup.find('input', {'name': 'currency2'})
    if usd_value_element:
        usd_value_text = usd_value_element['value']
        return usd_value_text
    else:
        return "USD Value not found."


```

```python

current_date, current_time = get_current_date_time()
usd_value = get_usd_value()
print("Data Atual:", current_date)
print("Hora Atual:", current_time)
print("Valor do Dólar:", usd_value)
```

```python
from pyspark import Row
df = spark.createDataFrame([
    Row(Date = current_date, Time = current_time, Price = usd_value)
])
```
